```python
'''Create a simple model for breast cancer dataset.

'''
import pandas as pd #importing the pandas and given an alias name as pd
import numpy as np
#STEP1:COLLECT THE DATA,RETRIEVE
data=pd.read_csv('C:/Users/Sahana R K/Desktop/SAHANA_ML_LAB/data_sets/Breast_Cancer.csv - Breast_Cancer.csv')#reading the file
#print(data) #statement to print the input file
#here we have 32 columns,DIAGNOSIS is the TARGET attribute

#STEP2:CLEAN THE DATA
#the ID column is unwanted here...so we will drop it
data.drop('id',axis=1,inplace=True)#by giving axis=1,we are specfying it as column,this will drop the column permanently as [inplace=true]
#print(data)#Now,the data set wont have ID Column

#STEP3:ANALYZE THE DATA
#In this case,it is seen that the target attribute[diagnosis]is already having binary values[B AND M],so we need to replace it.
#create a dictionary
#benign-0,Maligant-1
dict={'B':0,'M':1}
data.replace({'diagnosis':dict},inplace=True)
print(data)

#STEP4:WE SPLIT THE DATA TO TRAINING AND TESTING PHASE
#x-concept attributes,y-target attribute
x=data.iloc[:,2:31].values
#y=data.iloc[:(indicates that for all rows),0(the first column is stored in x)]
y=data.iloc[:,0].values

#NOW,WE SPLIT THEM TO TRAINING AND TESTING PHASE
#sklearn is the package,model selection is the model
from sklearn.model_selection import train_test_split as tt
x_train,x_test,y_train,y_test=tt(x,y)

#STEP5:TRAIN THE ALGORTHIM
#WE START TRAINING THEM,using logistic algorithm
#package:sklearn
#module:linear_model
#algorithm name:LogisticRegression
#and we give an alias name
from sklearn.linear_model import LogisticRegression as logr
model_logr=logr()#creating the object
#fit() is a function,parameters passed are the training data
model_logr.fit(x_train,y_train)# the system learns now

#Now,we test the data
#predict() is a function to predict the output,parameters passed are the testing data(1 variable)
pred_y=model_logr.predict(x_test)#testing the data
#print(y_test),this prints the actual output
#print(pred_y).this prints the predicted output

#We find the accuracy
#score() is function to find the accuracy of the testing data
logaccuracy=model_logr.score(x_test,y_test)
print("Accuracy:",logaccuracy)

#confusion matrix
#sklearn is the package,metrics is the model,and we import confusion matrix
from sklearn.metrics import confusion_matrix as cm
matrix=cm(y_test,pred_y)
print("Confusion Matrix")
print(matrix)#printing confusion matrix
```
